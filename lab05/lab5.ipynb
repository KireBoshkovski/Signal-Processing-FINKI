```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("SandP500.xlsx")

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'])
plt.title('Price over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
def weighted_moving_average(price, weight):
    weight_moving_avg = np.zeros_like(price)
    for i in range(len(price)):
        denominator = 0
        for j in range(weight):
            if i - j >= 0:
                weight_moving_avg[i] += (weight - j) * price[i - j]
                denominator += weight - j
        weight_moving_avg[i] /= denominator
    return weight_moving_avg
```

```python
# Weight
M = 3

# Plot both Price and Weight Moving Average
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price'], label='Original Price', color='blue')
plt.plot(df['Date'], weighted_moving_average(df['Price'], M), label=f'Weighted Moving Average (M={M})', color='red')
plt.plot(df['Date'], weighted_moving_average(df['Price'], M+7), label=f'Weighted Moving Average (M={M+7})', color='green')
plt.title('Price and Weighted Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
